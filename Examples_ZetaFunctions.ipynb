# Computing (local) Igusa and topological zeta functions for Newton-non-degenerated polynomials

We implement the formulas appearing in **[DH01]** and **[DL92]** for the computation of:

- the **Igusa's $p$-adic zeta function** $Z_{\text{Igusa}}(f;s)$,

- the **(Denef-Loeser) topological zeta function** $Z_{\text{top}}^{(d)}(f;s)$,

- their local versions $Z_{\text{Igusa},0}(f;s)$ and $Z_{\text{top},0}^{(d)}(f;s)$ at the origin,

(with variable $s\in\mathbb{C}$ verifying $\mathrm{Re}(s)>0$) for **Newton non-degenerated** polynomials and  holomorphic map germs $f$. We also implement a formula for the **monodromy zeta function** $\zeta_f(t)$ given by Varchenko in **[Var76]**.

## Settings

Let $f(\underline{x})=\sum_{\underline{\alpha}\in\mathbb{N}^n}a_{\underline{\alpha}} x_1^{\alpha_1}\cdots x_n^{\alpha_n}$ be a non-constant polynomial (resp. holomorphic map germ at 0). Set the following:

- *Support*: $\text{supp}(f) = \{\underline{\alpha}\in\mathbb{N}^n \mid a_{\underline{\alpha}}\neq0\}$,

- *Newton polyhedron*: $\Gamma(f) = \text{Conv}\left(\bigcup_{\underline{\alpha}\in \text{supp}(f)} \underline{\alpha} + (\mathbb{R}_{\geq0})^n\right) $,

- For any face $\tau\in\Gamma(f)$: $f_\tau := \sum_{\underline{\alpha}\in\tau} a_{\underline{\alpha}} x_1^{\alpha_1}\cdots x_n^{\alpha_n}$.

- $f$ is *Newton non-degenerated over a field $k$* if for any closed (resp. compact) face $\tau\in\Gamma(f)$ the system $f_\tau = \partial_{x_1} f_\tau = \cdots \partial_{x_m} f_\tau = 0$ has no solutions on $(k^\ast)^n$, 



## Igusa's $p$-adic zeta function

Let $f$ be in $\mathbb{Z}_p\left[x_1,\ldots,x_n\right]$. The *global* (resp. *local*) *Igusa zeta function associated to $f$* is the function in $\mathbb{Q}(p^{-s})$ defined as

$$
Z_{\text{Igusa}}(f;s) = \int_{\mathbb{Z}_p^n} |f(\underline{x})|_p^s\, |\mathrm{d}\underline{x}|,\qquad\left(\text{resp. }\ Z_{\text{Igusa},0}(f;s) = \int_{p\mathbb{Z}_p^n} |f(\mathbf{x})|_p^s\, |\mathrm{d}\underline{x}|\right),\qquad s\in\mathbb{C},\ \mathrm{Re}(s)>0,
$$

where $|\mathrm{d}\underline{x}|$ denotes the $n$-dimensional Haar measure in $\mathbb{Q}_p^n$ normalized such as $\mathbb{Z}_p^n$ has measure 1.

### Formula of Igusa's $p$-adic zeta function for the Newton-non-degenerated case

Let $p>1$ be a prime number and define the following:

- $N_\tau := \#\lbrace a\in (\mathbb{F}_p^\times)^n\ |\ \bar{f}_\tau(a)=0\rbrace$,
- $L_\tau (s) := p^{-n}\left((p-1)^n - p N_\tau \frac{p^s-1}{p^{s+1}-1}\right)$,
- $S_{\Delta_\tau} (s) := \sum_{\underline{k}\in\mathbb{N}^n\cap\Delta_\tau} p^{ - m(\underline{k}) s -\sigma (\underline{k})}$, where $\Delta_\tau$ is the dual cone of $\tau$, with $\sigma(\underline{k}) = \sum_i k_i$ and $m(\underline{k}) = \min\{ \underline{k}\cdot v \mid v \text{ vertex of } \Gamma(f) \}.$

If $f$ is Newton no-degenerated over $\mathbb{F}_p$ , then

$$
Z_{\text{Igusa}}(f;s) = L_{\Gamma(f)}(s) + \sum_{\substack{\tau\ \text{proper face}\\ \text{of}\ \Gamma(f)}} L_\tau(s) S_{\Delta_\tau}(s)\qquad \left(\text{resp. \quad } Z_{\text{Igusa},0}(f;s) = \sum_{\substack{\tau\ \text{compact face}\\ \text{of}\ \Gamma(f)}} L_\tau(s) S_{\Delta_\tau}(s).\right).
$$

## The (Denef-Loeser) topological zeta function

Let $f$ be a non-constant complex polynomial (resp. holomorphic map germ) and the canonical volume form $\omega_0 = dx_1\wedge\cdots\wedge dx_n$. Let $\pi: X \to \mathbb{C}^n$ be an embedded resolution of $f$ with simple normal crossings and consider:

$$
\text{div}(\pi^\ast f) = \sum_{i\in J} N_i E_i\qquad\text{and}\qquad \text{div}(\pi^\ast\omega_0 ) = \sum_{i\in J} (\nu_i-1) E_i.
$$

Taking the stratification of $X$ by the smooth parts $\mathring{E}_I = \left(\cap_{i\in I} E_i\right)\setminus \left(\cup_{j\not\in I} E_j\right)$ and a natural number $d\geq1$, the *global (resp. local) topological zeta function associated to $f$* is the element in $\mathbb{Q}(s)$ given by

$$
Z_{\text{top}}^{(d)}(f;s) = \sum_{\substack{I\subset J\\\forall i\in I,\ d|N_i}} \chi(\mathring{E}_I)\prod_{i\in I}\frac{1}{N_i s + \nu_i}\qquad\left(\text{resp. }\quad Z_{\text{top},0}^{(d)}(f;s) = \sum_{\substack{I\subset J\\\forall i\in I,\ d|N_i}} \chi(\mathring{E}_I\cap\pi^{-1}(0))\prod_{i\in I}\frac{1}{N_i s + \nu_i}\right)
$$

### Formula of Ithe topological zeta function for the Newton-non-degenerated case

Define the following:

- Multiplicity of a simple cone $\Delta$:   $\text{mult}(\Delta):=\# \mathbb{N}^r\cap \{\lambda_1a_1+\cdots+\lambda_ra_r \mid \lambda_i\in [0,1)\}$, where $a_1,\ldots,a_r$ are primitive vectors in $\mathbb{N}^n$ generating $\Delta$.

- $J_{\Delta}(s) := \frac{\text{mult}(\Delta_i)}{( m(a_{1})s + \sigma(a_{1}))\cdots (m(a_{r})s + \sigma(a_{r}))}$.

- For proper face $\tau\in\Gamma(f)$:  $J(\tau,s) := \sum_{i=1}^r J_{\Delta_i}$ , decomponing  $\Delta_\tau = \bigcup_{i=1}^r \Delta_i$ in simple cones of dimension $\dim\Delta_\tau = r$ such that $\dim (\Delta_i\cap\Delta_j) < l, \forall i\neq j$. 

- $J(\Gamma(f),s):=1$.

If $f$ is Newton no-degenerated over $\mathbb{C}$ , then

$$
Z_{\text{top}}(f;s) = \sum_{\substack{\tau\ \text{vertex}\\ \text{of}\ \Gamma(f)}} J(\tau,s) + \left(\frac{s}{s+1}\right)\sum_{\substack{\tau\ \text{face of}\ \Gamma(f)\\ \dim\tau\geq 1}} (-1)^{\dim\tau}(\dim\tau)!\text{Vol}(\tau)J(\tau,s),
$$

$$
Z_{\text{top}}^{(d)}(f;s) = \sum_{\substack{\tau\ \text{face of}\ \Gamma(f)\\ d|m(\Delta_\tau)}} (-1)^{\dim\tau}(\dim\tau)!\text{Vol}(\tau)J(\tau,s),\quad\text{for}\ d>1.
$$

Respectively, in the local case one can compute $Z_{\text{top},0}(f;s)$ and $Z_{\text{top},0}^{(d)}(f;s)$ by the same formulas but considering only compact faces.


## Notes

* The monomdromy zeta function $\zeta_f(t)$ is also computed in the local Newton no-degenerated case following **[Var76]**.

* The following examples are taken from the original work of K. Hoornaert and D. Loots in **[HL00]**.

* One can generalize the above formulas whenever we consider a monomial volume form $\omega = x_1^{w_1-1}\cdots x_n^{w_n-1}\cdot \omega_0$ instead the canonical volume form, defining $\sigma(\underline{k}) = \sum_i k_iw_i$.

## **Examples** computing the **Igusa zeta function**

### Example 1: $f(x,y,z)=x^2 - y^2 + z^3$

```python
R.<x,y,z> = QQ[]
zex1 = ZetaFunctions(x^2 - y^2 + z^3)
```

```python
zex1.give_info_newton(faces = True)
```

```python
zex1.newton_plot()
```

```python
zex1.cones_plot()
```

For $p=3$:

```python
zex1.igusa_zeta(p=3)
```

```python
show(_)
```

For an arbitrary $p$, without information over the faces:

```python
zex1.igusa_zeta()
```

```python
show(_)
```

For an arbitrary $p$, if one has the number of solutions over the faces encoded by a _dictionary_:

```python
dNtau3 = { x^2-y^2+z^3 : (p-1)*(p-3), -y^2+z^3 : (p-1)^2, x^2+z^3 : (p-1)^2, x^2-y^2 : 2*(p-1)^2 }
```

```python
zex1.igusa_zeta(dict_Ntau = dNtau3)
```

```python
show(_)
```
